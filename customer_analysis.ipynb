**Kurtosis**
- kurtosis is a statistical measure that describe tha shape of a distribution to its peak.
- It provide insights into the "tailedness" or extremely of data points.

**Types of Kurtosis**
- Mesokurtic (kurtosis = 3) : The distribution has a normal shape.
- leptokurtic (kurtosis > 3) : The distribution has heavy tails and a sharp peak and it indicate more extreme outlier than a normal distribution.
- platykurtic (kurtosis < 3) : The distribution has light tails and a flatter peak and fewer extreme outliers compared to normal distribution.

**Skewness**
- Skewness measures the asymmetry of a distribution.
- It can indicate whether the data is symmetric or skewed to the left or right.

**Types of skewness**
- Skewness = 0 : Perfectly symmetric distribution (eg. normal distribution).
- Positive skewness (>0) : The tail on the right side is longer, and the bulk of data is concentrated on the left.
- Negative skewness (<0) : The tail on the left side is longer, and the bulk of data is concentrated on the right.

# ANNOVA

**Definition** :
- ANNOVA stands for Analysis of Variance.
- ANNOVA is a statistical test used to compare the means of two or more groups and determine if they are significantly different from each other.
- It can apply only categorical data.
- It can calculated by two methods:
- 1.F-statistics : The ratio of variance between group to variance within groups.
- 2.p-value : It is the probability of observing data to true differences between the group. If p-value < 0.05 i.e. we reject the null hypothesis, if p-value > 0.05 then

**Types of ANNOVA**

1.ONE-WAY :
- There is only one independent variable and with more thean two levels or groups.

2.TWO-WAY :
- Two-way ANNOVA is two different categorical independent variables on one dependent variables.

```python
from scipy.stats import f_oneway 

cust_cat_features = ['Product_Interest']
for feature in cust_cat_features:
    groups = [cust[cust[feature] == category]['Annual_Spending'] for category in cust[feature].unique()]
    f_stat, p_value = f_oneway(*groups) #unpacks the list of groups into separate arguments for the function.
    
    print(f"ANOVA results for {feature}:")
    print(f" F-statistic: {f_stat:.2f}")
    print(f" P-value: {p_value:.4f}\n")
```

**Product_Interest**
- f-statistic = 1.12
- p-value = 0.3376. The p-value is greater than 0.05 means the product interest has no significant effect and we fail to reject the null hypothesis.


**Formulated Hypothesis**:
- Hypothesis 1 : Higher loyalty score and online activity levels are associated with higher spending.
- Hypothesis 2 : Customer with shorter recency periods are more likely to purchase.
- Hypothesis 3 : Income category has a significant influence on purchase likelihood.
- Hypothesis 4 : Product interest is a key factor influencing annual spending.

```python
# Hypothesis 1 : online activity level vs customer loyalty score
from scipy.stats import f_oneway # performs a one-way ANNOVA test for only one independent variable.
# define categorical columns
cust_cat_features = ['Customer_Loyalty_Score','Online_Activity_Level']
for feature in cust_cat_features:
    groups = [cust[cust[feature] == category]['Annual_Spending'] for category in cust[feature].unique()]
    f_stat, p_value = f_oneway(*groups) #unpacks the list of groups into separate arguments for the function.
    
    print(f"ANOVA results for {feature}:")
    print(f" F-statistic: {f_stat:.2f}")
    print(f" P-value: {p_value:.4f}\n")
```

**Customer Loyalty Score**
- p-value is 0.0000 means value is less than 0.05 then reject null hypothesis.

**Online activity level**
- p-value is 0.0741 means value is greater than 0.05 then fail to reject null hypothesis.

# Chi Square Test

- checking if two categorical variables are related or not.
- comparing observed vs expected frequencies.
- If there is a big difference what is expected and what observed that suggest the two variable might be.

```python
import pandas as pd
from scipy.stats import chi2_contingency # contingency use for calculate the statistical significance of the relationship

# perform chi square test on two or more categorical variables
contingency_table = pd.crosstab(cust['Income_Category'],cust['Will_Purchase']) # crosstab fro comparing two variable
chi2_test,p,Dof,expected = chi2_contingency(contingency_table)
print("p-value:",p) # check value is greater than 0.05 or not
if p<0.05:
    print("The variable are realted")
else:
    print("The variable are not related")

print("chi square test:",chi2_test)
print("Degree of freedom:",Dof) # Dof is number of independent values
print("Expected frequancy:\n",expected)
```
